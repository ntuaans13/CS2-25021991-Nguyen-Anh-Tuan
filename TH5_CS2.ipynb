```python
#W5A1
def Max_Of_Two(a, b): 
    return a if a > b else b

a, b = map(int, input().split())
print(Max_Of_Two(a, b))
```

```python
#W5A2
def Swap(a, b): 
    return [b, a]

a, b = map(int, input().split())
a, b = Swap(a, b)
print(a, b)
```

```python
#W5A3 
def isPrime(n):
    for i in range(2, int(n ** 0.5 + 1)):
        if(n % i == 0):
            return False
    return n > 1

n = int(input())
print("YES" if isPrime(n) else "NO")
```

```python
#W5A9
def Sum_of_Digits(n):
    sum = 0
    while(n):
        sum += n % 10
        n //= 10
    
    return sum

n = int(input())
print(Sum_of_Digits(n))
```
